```python
i = 0
final_score = 0
random_value = 0
import warnings
warnings.filterwarnings('ignore')
while True:
    X_train, X_test, y_train, y_test = train_test_split(wine.data,wine.target,test_size=0.2,random_state=i)
    # model creation
    dt_model = DecisionTreeClassifier()
    dt_model.fit(X_train,y_train)
    
    score = dt_model.score(X_test,y_test)
    
    if score>final_score:
        final_score = score
        random_value = i
    print(f'Iterations: {i}, Achieved Score: {final_score}, Best random value: {random_value}')
    display(clear=True)
    if score == 1:
        break
    i+=1
```
