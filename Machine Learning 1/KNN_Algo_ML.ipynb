```python
import time
m_score =  0 
k = 1
k_value = []
model_score  = []
while True:
    X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=200,test_size=0.2)

    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train,y_train)
    score = model.score(X_test,y_test)  
    
    if m_score < score:
        m_score = score
    
    print(f'K value: {k}',score)
    k+=1
    
    k_value.append(k)
    model_score.append(score)
    if k ==15:
        break
    display(clear=True)
#     time.sleep(2)
```

```python
import numpy as np
```

```python
plt.plot(k_value,model_score)
plt.xlabel('K Value')
plt.ylabel('Model Score')

array_score = np.array(model_score)
highest_score = max(array_score)
highest_score_index = np.argmax(array_score)
k_value_index = k_value[highest_score_index]

plt.annotate(f"Highest Score {round(highest_score*100,2)}\n  At value of K = {k_value_index}", xy=(k_value_index, highest_score), xytext=(10,0.8),arrowprops=dict(arrowstyle="->"))
plt.show()
```
